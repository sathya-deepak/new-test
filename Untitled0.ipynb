```python
import random

# Simulated bus schedule data
bus_schedule = {
    "madurai": ["8:00 AM", "12:00 PM", "4:00 PM"],
    "chennai": ["9:30 AM", "1:30 PM", "5:30 PM"],
    "virudhnagar": ["10:15 AM", "2:15 PM", "6:15 PM"]
}

def get_simulated_bus_timings(source, destination):
    if source in bus_schedule and destination in bus_schedule:
        timings = random.choice(bus_schedule[source])
        return f"Next bus from {source} to {destination} departs at {timings}"
    else:
        return "Bus timings are not available for the provided route."

# Get input from the user
source = input("Enter the source place: ")
destination = input("Enter the destination place: ")

# Get and print simulated bus timings
bus_timings = get_simulated_bus_timings(source, destination)
print("Bot:", bus_timings)
```
